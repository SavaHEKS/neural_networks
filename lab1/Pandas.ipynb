```python
import pandas as pd
import numpy as np
```

# Creating DataFrames :

### Create from data recorded in the program code


```python
df1 = pd.DataFrame({
    'Name' : ['Sava' , 'Vika' , 'Felix'] ,
    'Surname' : ['Ivanina' , 'Soroka' , 'Petrus'],
    'Age' : [20 , 21 , 22]
})
df1
```

### Create from numpy ndarray:


```python
df2 = pd.DataFrame(np.array([['Sava', 'Ivanina', 20], ['Vika', 'Soroka', 21], ['Felix', 'Petrus', 22]]),
                   columns=['Name', 'Surname', 'Age'])
df2
```

### Create from dict:

```python
dict_clients = {'Name': ['Sava', 'Vika', 'Felix'], 'Surname' : ['Ivanina', 'Soroka' , 'Petrus'], 'Age' : [20 , 21 , 22]}
df3 = pd.DataFrame(dict_clients)
df3
```

### Create from Website's table

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_music_artists'
df4 = pd.read_html(url)[0]
df4
```
