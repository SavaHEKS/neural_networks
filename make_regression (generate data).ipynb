```python
from sklearn.datasets import make_regression
import pandas as pd
```

```python
 X, Y = make_regression(n_samples=250000,
                              n_features=25,
                              n_informative=10,
                              n_targets=1,
                              bias = 10.00,
                              effective_rank = 3,
                              tail_strength=0.3,
                              noise=0.02,
                              shuffle=True,
                              coef=False,
                              random_state=666)
```

```python
df = pd.DataFrame(X)
df = df.add_prefix('x')
df['y'] = Y
df
```

```python
df.to_csv('data.csv', sep=',', index=False)
```
