```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('data.csv', sep=',')
df.head(3)
```

```python
X_df = df.iloc[:,:-1]
X_df.head(3)
```

```python
y_df = df.iloc[:,-1]
y_df.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, train_size=0.70, random_state=50)
```

```python
model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", model.score(X_test, y_test))
print(y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
figure(figsize=(25, 13))
X_graph = np.arange(len(y_test))
plt.scatter(X_graph, y_test, color = 'yellow')
plt.scatter(X_graph, y_pred, color = 'black')
ax = plt.axes()
ax.set_facecolor("white")
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```
